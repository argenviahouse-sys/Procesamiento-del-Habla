```python
# 4. Carga del LLM (google/flan-t5-small)
print("Cargando LLM: google/flan-t5-small...")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
model_llm = AutoModelForCausalLM.from_pretrained("google/flan-t5-small")
llm_pipeline = pipeline(
    "text2text-generation",
    model=model_llm,
    tokenizer=tokenizer
)
print("LLM cargado exitosamente.")

# 5. ConfiguraciГіn de Embedding Models adicionales para ChromaDB

# --- Gensim GloVe Embedding Function para ChromaDB ---
# Cargamos el modelo GloVe pre-entrenado de Gensim
print("Cargando modelo GloVe (glove-wiki-gigaword-50) para Gensim...")
glove_model = api.load("glove-wiki-gigaword-50")
print("Modelo GloVe cargado exitosamente.")

# Clase para la funciГіn de embedding de ChromaDB usando GloVe
class GensimEmbeddingFunction(embedding_functions.EmbeddingFunction):
    def __init__(self, model):
        self.model = model

    def __call__(self, input_texts):
        embeddings = []
        for text in input_texts:
            # Tokenizar el texto y obtener vectores de palabras
            words = text.lower().split()
            # Filtrar palabras que no estГЎn en el vocabulario de GloVe
            word_vectors = [self.model[word] for word in words if word in self.model.key_to_index]
            if word_vectors:
                # Promediar los vectores de palabras para obtener el embedding de la oraciГіn
                embeddings.append(np.mean(word_vectors, axis=0))
            else:
                # Si no hay palabras en el vocabulario, usar un vector de ceros
                embeddings.append(np.zeros(self.model.vector_size))
        return embeddings

gensim_ef = GensimEmbeddingFunction(glove_model)

# 6. CreaciГіn y poblaciГіn de la colecciГіn ChromaDB para GloVe
print("Creando y poblando colecciГіn ChromaDB con GloVe embeddings...")
client_glove = chromadb.Client()
coleccion_glove = client_glove.get_or_create_collection(name="conocimiento_bancario_glove", embedding_function=gensim_ef)

# Solo agregar documentos si la colecciГіn estГЎ vacГ­a para evitar duplicados
if coleccion_glove.count() == 0:
    documentos_glove = [item["pregunta"] for item in dataset_evaluacion]
    metadatos_glove = [{"respuesta": item["respuesta"]} for item in dataset_evaluacion]
    ids_glove = [str(i) for i in range(len(dataset_evaluacion))]

    coleccion_glove.add(
        documents=documentos_glove,
        metadatas=metadatos_glove,
        ids=ids_glove
    )
    print(f"ColecciГіn 'conocimiento_bancario_glove' creada y poblada con {coleccion_glove.count()} documentos.")
else:
    print(f"ColecciГіn 'conocimiento_bancario_glove' ya contiene documentos. Saltando operaciГіn de aГ±adir. Documentos: {coleccion_glove.count()}.")
```

```python
# InstalaciГіn de las bases de datos vectoriales recomendadas
!pip install chromadb --quiet
!pip install faiss-cpu --quiet

import chromadb
from chromadb.utils import embedding_functions

# 1. Configurar el cliente y crear una colecciГіn (tu base de conocimientos)
client = chromadb.Client()
# Usamos un modelo de embedding por defecto o el que hayas elegido (ej. Sentence Transformers)
embedding_fn = embedding_functions.DefaultEmbeddingFunction()
coleccion = client.get_or_create_collection(name="conocimiento_bancario", embedding_function=embedding_fn)

# 2. Preparar los datos de tu dataset (Fase 1 y TP4)
# Extraemos las preguntas para vectorizar y las respuestas como 'metadatos'
documentos = [item["pregunta"] for item in dataset_evaluacion]
metadatos = [{"respuesta": item["respuesta"]} for item in dataset_evaluacion]
ids = [str(i) for i in range(len(dataset_evaluacion))]

# 3. Cargar el dataset en la base de datos vectorial
# Only add documents if the collection is empty to avoid duplicates on re-execution.
if coleccion.count() == 0:
    coleccion.add(
        documents=documentos,
        metadatas=metadatos,
        ids=ids
    )
else:
    print("Collection already contains documents. Skipping add operation.")

# 4. FunciГіn de bГєsqueda para tu ChatBot (Motor de bГєsqueda)
def buscar_en_db(consulta_usuario):
    # La DB vectoriza la consulta y busca la similitud del coseno automГЎticamente [8, 9]
    resultados = coleccion.query(
        query_texts=[consulta_usuario],
        n_results=1
    )
    # Retornamos la respuesta recuperada
    # The 'metadatas' field from ChromaDB query results is a list of lists.
    # We need to access the first element of the outer list, then the first element
    # of the inner list to get the metadata dictionary, and finally the 'respuesta' key.
    return resultados['metadatas'][0][0]['respuesta']

# Prueba rГЎpida
print(f"Chatbot: {buscar_en_db('ВїCГіmo recupero mi clave?')}")

# Bucle infinito para mantener la conversaciГіn activa
while True:
    # 1. Solicitar la entrada del usuario
    entrada_usuario = input("TГє: ")

    # 2. CondiciГіn de salida para cerrar el chat
    if entrada_usuario.lower() in ["salir", "chau", "exit"]:
        print("Chatbot: ВЎHasta luego! Gracias por consultar al Banco Nacional.")
        break

    # 3. Llamar a tu funciГіn de bГєsqueda con la pregunta ingresada
    # AquГ­ reemplazamos la consulta fija por 'entrada_usuario'
    respuesta = buscar_en_db(entrada_usuario)

    # 4. Mostrar la respuesta recuperada
    print(f"Chatbot: {respuesta}")
```

```python
# 7. DefiniciГіn de la clase ChatBot (Consigna c)
class ChatBot:
    def __init__(self, llm_pipeline, db_collection):
        self.llm_pipeline = llm_pipeline
        self.db_collection = db_collection

    def responder(self, query):
        # 1. Recuperar contexto relevante de la base de datos vectorial
        # La colecciГіn de ChromaDB se encarga de embeddear la query con su funciГіn de embedding configurada
        resultados = self.db_collection.query(
            query_texts=[query],
            n_results=1
        )

        # Extraer el contexto (respuesta mГЎs similar)
        contexto = resultados['metadatas'][0][0]['respuesta']

        # 2. Generar respuesta usando el LLM y el contexto recuperado
        # Formato de prompt para el LLM
        prompt = f"Basado en la siguiente informaciГіn: '{contexto}', responde a la pregunta: '{query}'."

        # Generar la respuesta con el LLM
        # Ajustamos parГЎmetros para una generaciГіn mГЎs controlada
        response = self.llm_pipeline(
            prompt,
            max_new_tokens=50,      # Limita la longitud de la respuesta
            num_return_sequences=1, # Solo queremos una respuesta
            truncation=True         # Trunca el prompt si es demasiado largo para el modelo
        )
        return response[0]['generated_text']

# 8. InstanciaciГіn de los ChatBots para comparaciГіn (Consigna d)
# bot_st: Usa la colecciГіn 'conocimiento_bancario' (creada en la celda anterior,
#         que utiliza la DefaultEmbeddingFunction de ChromaDB, normalmente SentenceTransformer)
#         Adaptamos el nombre 'bot_spacy' a 'bot_st' para mayor claridad con el modelo de embedding usado.
bot_st = ChatBot(llm_pipeline, coleccion)

# bot_gensim: Usa la colecciГіn 'conocimiento_bancario_glove' (creada con Gensim GloVe embeddings)
bot_gensim = ChatBot(llm_pipeline, coleccion_glove)

print("ChatBots 'bot_st' y 'bot_gensim' instanciados exitosamente.")
```

```python
revisar y analizar mi pipeline para el chatbot, verificar si los pasos estan bien organizados, dar recomendaciones, alternativas y corregir cuando es necesario.

#1
preguntas = ["ВїCГіmo abro una caja de ahorro?",
             "ВїCuГЎles son los requisitos para un prГ©stamo personal?",
             "PerdГ­ mi tarjeta de dГ©bito, ВїquГ© hago?",
             "ВїCГіmo consulto mi saldo?",
             "ВїQuГ© es el CBU y dГіnde lo encuentro?",
             "ВїCГіmo constituyo un plazo fijo?",
             "ВїCuГЎl es el horario de atenciГіn en sucursales?",
             "ВїCГіmo solicito una tarjeta de crГ©dito?",
             "ВїPuedo retirar dГіlares por cajero automГЎtico?",
             "El cajero no me dio el dinero pero me lo descontГі, ВїcГіmo reclamo?",
             "ВїCГіmo pago mis impuestos y servicios?",
             "ВїCГіmo activo mi tarjeta de dГ©bito nueva?",
             "ВїTienen crГ©ditos hipotecarios disponibles?",
             "OlvidГ© mi clave PIN del cajero, ВїcГіmo la blanqueo?",
             "ВїCuГЎl es el lГ­mite diario de extracciГіn?",
             "ВїCГіmo realizo una transferencia a otro banco?",
             "ВїQuГ© es el Alias CBU?",
             "ВїCГіmo saco un turno para ir a la sucursal?",
             "ВїOfrecen seguros para el hogar?",
             "ВїCГіmo me registro en la App del banco?"]

# Listado de respuestas correspondientes
respuestas = ["Puedes abrirla 100% online desde nuestra app con tu DNI o de forma presencial en cualquier sucursal.",
              "Debes ser mayor de 18 aГ±os, contar con ingresos mГ­nimos demostrables y tener una antigГјedad laboral de 6 meses.",
              "Denunciala inmediatamente por Home Banking o llamando al centro de atenciГіn para bloquearla y solicitar la reposiciГіn.",
              "Puedes hacerlo gratis por la App mГіvil, el Home Banking o en cualquier cajero automГЎtico de la red.",
              "Es tu Clave Bancaria Uniforme. Lo encontrГЎs en la secciГіn 'Detalle de cuenta' de tu App o Home Banking.",
              "Ingresa a la secciГіn de 'Inversiones' en la App, elegГ­ el monto y el plazo (mГ­nimo 30 dГ­as) y confirma la operaciГіn",
              "Nuestras sucursales atienden de lunes a viernes en el horario bancario habitual de 10:00 a 15:00 horas.",
              "Puedes pedirla online desde la web oficial adjuntando tu Гєltimo recibo de sueldo o en tu sucursal mГЎs cercana.",
              "SГ­, solo en cajeros habilitados para operar con moneda extranjera y si posees una caja de ahorro en dГіlares.",
              "Debes realizar el reclamo telefГіnico al nГєmero que figura en el cajero o comunicarte con nuestro soporte tГ©cnico.",
              "Puedes hacerlo desde la opciГіn 'Pagos' en el Home Banking o escaneando el cГіdigo de barras desde nuestra App.",
              "Simplemente realiza una consulta de saldo o una extracciГіn en cualquier cajero automГЎtico con tu clave provisoria.",
              "SГ­, contamos con lГ­neas vigentes para compra, construcciГіn o refacciГіn de vivienda Гєnica. ConsultГЎ las tasas online.",
              "Puedes realizar el blanqueo de PIN directamente desde el Home Banking, en la secciГіn 'GestiГіn de Tarjetas'.",
              "El lГ­mite estГЎndar se puede consultar y modificar desde tu Home Banking en la configuraciГіn de perfil.",
              "Desde la secciГіn 'Transferencias', ingresa el CBU o Alias del destinatario, el monto y confirma con tu clave de seguridad.",
              "Es un nombre corto y fГЎcil de recordar que reemplaza al nГєmero de CBU para facilitar tus transferencias.",
              "Solicita tu turno previo desde nuestro sitio web oficial eligiendo la sucursal, el dГ­a y el trГЎmite a realizar.",
              "SГ­, disponemos de coberturas contra incendio, robo y daГ±os por agua para proteger tu vivienda y bienes.",
              "DescargГЎ la aplicaciГіn oficial, selecciona 'Registrarse' y seguГ­ los pasos de validaciГіn biomГ©trica con tu cГЎmara."]



print("las preguntas son:", preguntas)
print("las respuestas son:", respuestas)

#2
dataset_evaluacion = [
    {"pregunta": "ВїCГіmo puedo enviar dinero al exterior?",
     "respuesta": "Debes realizar una transferencia internacional desde la secciГіn de Comex en Home Banking, adjuntando la documentaciГіn requerida."},

    {"pregunta": "No recuerdo mi clave de Home Banking, Вїla recupero?",
     "respuesta": "Selecciona 'OlvidГ© mi clave' en la pantalla de inicio y valida tu identidad mediante el cГіdigo enviado a tu celular o mail."},

    {"pregunta": "ВїQuГ© seguros ofrecen para vehГ­culos?",
     "respuesta": "Contamos con coberturas de Terceros Completo y Todo Riesgo con diferentes opciones de franquicia para tu auto."},

    {"pregunta": "ВїCГіmo solicito una chequera nueva?",
     "respuesta": "Puedes pedirla desde la gestiГіn de cuentas en Home Banking o retirarla en la sucursal donde estГЎ radicada tu cuenta."},

    {"pregunta": "ВїCuГЎl es la tasa de interГ©s actual del plazo fijo?",
     "respuesta": "Las tasas vigentes se actualizan diariamente y puedes consultarlas en la secciГіn 'Inversiones' de nuestra web oficial."},

    {"pregunta": "Necesito aumentar el lГ­mite de mi tarjeta de crГ©dito.",
     "respuesta": "La solicitud se realiza por la App enviando tu Гєltimo comprobante de ingresos para una evaluaciГіn crediticia."},

    {"pregunta": "ВїTienen cajas de seguridad disponibles?",
     "respuesta": "La disponibilidad depende de cada sucursal. Te recomendamos consultar presencialmente en la sede mГЎs cercana."},

    {"pregunta": "ВїCГіmo pago el resumen de mi tarjeta de crГ©dito?",
     "respuesta": "Puedes pagarlo por Home Banking, cajero automГЎtico o mediante dГ©bito automГЎtico configurando tu cuenta."},

    {"pregunta": "ВїA cuГЎnto estГЎ la cotizaciГіn del dГіlar hoy?",
     "respuesta": "Puedes ver el valor de compra y venta actualizado al instante ingresando a tu cuenta en la App o Home Banking."},

    {"pregunta": "CambiГ© mi domicilio, ВїcГіmo actualizo mis datos?",
     "respuesta": "Ingresa a 'Perfil' en la App y sube una foto de tu nuevo DNI o un servicio a tu nombre para validar el cambio."},

    {"pregunta": "ВїQuГ© beneficios tengo por cobrar mi sueldo aquГ­?",
     "respuesta": "Accedes a una cuenta sueldo sin costos de mantenimiento, tasas preferenciales en prГ©stamos y descuentos en comercios."},

    {"pregunta": "ВїCГіmo utilizo el Token de seguridad en la App?",
     "respuesta": "El Token se activa por Гєnica vez en un cajero y sirve para autorizar transferencias y pagos de forma segura."},

    {"pregunta": "Veo un movimiento sospechoso en mi cuenta, ВїquГ© hago?",
     "respuesta": "Desconoce el cargo inmediatamente desde la App en la secciГіn de movimientos o llama a nuestro centro de denuncias."},

    {"pregunta": "ВїOfrecen lГ­neas de crГ©dito para emprendedores?",
     "respuesta": "SГ­, contamos con prГ©stamos productivos para MiPyMEs con plazos de gracia y tasas subsidiadas por el Estado."},

    {"pregunta": "ВїPuedo abrir una cuenta para mi hijo menor de edad?",
     "respuesta": "SГ­, ofrecemos la 'Cuenta Adolescente' que permite a menores de 13 a 17 aГ±os gestionar sus ahorros con supervisiГіn."},

    {"pregunta": "ВїCГіmo retiro mi jubilaciГіn si no tengo la tarjeta?",
     "respuesta": "Puedes generar un cГіdigo de extracciГіn sin tarjeta desde la App o presentarte en ventanilla con tu DNI fГ­sico."},

    {"pregunta": "El cajero se quedГі con mi tarjeta, ВїcГіmo la recupero?",
     "respuesta": "Por seguridad, el banco la destruye. Debes pedir una reposiciГіn por Home Banking y te llegarГЎ a tu domicilio en 72 horas."},

    {"pregunta": "ВїTienen algГєn programa de puntos por consumos?",
     "respuesta": "SГ­, con tus compras sumas puntos en nuestro programa de beneficios para canjear por premios o descuentos."},

    {"pregunta": "ВїCГіmo pido un prГ©stamo prendario para un auto?",
     "respuesta": "Inicia la solicitud online cargando los datos del vehГ­culo y recibirГЎs una oferta de financiaciГіn en el acto."},

    {"pregunta": "Deseo cerrar mi cuenta bancaria definitivamente.",
     "respuesta": "Puedes solicitar la baja total desde Home Banking siempre que la cuenta no presente deudas o saldo negativo."}
]

print(f"Dataset de evaluaciГіn cargado con {len(dataset_evaluacion)} pares de preguntas y respuestas.")

#3
!pip install spacy gensim transformers accelerate chromadb scikit-learn --quiet
# Descarga del modelo de lenguaje medio para SpaCy con vectores reales [4, 5]
!python -m spacy download es_core_news_md --quiet

# 2. IMPORTACIONES
import spacy
import numpy as np
import pandas as pd
import gensim.downloader as api
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from google.colab import userdata
from huggingface_hub import login
from sklearn.metrics.pairwise import cosine_similarity

# 3. AUTENTICACIГ“N SEGURA (Consigna b)
# Se recupera el token de la pestaГ±a 'Secrets' de Colab [ConversaciГіn previa]
hf_token = userdata.get('HF_TOKEN')
if hf_token:
    login(token=hf_token)
    print("AutenticaciГіn exitosa con Hugging Face.")
else:
    print("ADVERTENCIA: No se encontrГі HF_TOKEN. Algunos modelos LLM podrГ­an no cargar.")

#4
print("Cargando LLM: google/flan-t5-small...")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
model_llm = AutoModelForCausalLM.from_pretrained("google/flan-t5-small")
llm_pipeline = pipeline(
    "text2text-generation",
    model=model_llm,
    tokenizer=tokenizer
)
print("LLM cargado exitosamente.")

# 5. ConfiguraciГіn de Embedding Models adicionales para ChromaDB

# --- Gensim GloVe Embedding Function para ChromaDB ---
# Cargamos el modelo GloVe pre-entrenado de Gensim
print("Cargando modelo GloVe (glove-wiki-gigaword-50) para Gensim...")
glove_model = api.load("glove-wiki-gigaword-50")
print("Modelo GloVe cargado exitosamente.")

# Clase para la funciГіn de embedding de ChromaDB usando GloVe
class GensimEmbeddingFunction(embedding_functions.EmbeddingFunction):
    def __init__(self, model):
        self.model = model

    def __call__(self, input_texts):
        embeddings = []
        for text in input_texts:
            # Tokenizar el texto y obtener vectores de palabras
            words = text.lower().split()
            # Filtrar palabras que no estГЎn en el vocabulario de GloVe
            word_vectors = [self.model[word] for word in words if word in self.model.key_to_index]
            if word_vectors:
                # Promediar los vectores de palabras para obtener el embedding de la oraciГіn
                embeddings.append(np.mean(word_vectors, axis=0))
            else:
                # Si no hay palabras en el vocabulario, usar un vector de ceros
                embeddings.append(np.zeros(self.model.vector_size))
        return embeddings

gensim_ef = GensimEmbeddingFunction(glove_model)

# 6. CreaciГіn y poblaciГіn de la colecciГіn ChromaDB para GloVe
print("Creando y poblando colecciГіn ChromaDB con GloVe embeddings...")
client_glove = chromadb.Client()
coleccion_glove = client_glove.get_or_create_collection(name="conocimiento_bancario_glove", embedding_function=gensim_ef)

# Solo agregar documentos si la colecciГіn estГЎ vacГ­a para evitar duplicados
if coleccion_glove.count() == 0:
    documentos_glove = [item["pregunta"] for item in dataset_evaluacion]
    metadatos_glove = [{"respuesta": item["respuesta"]} for item in dataset_evaluacion]
    ids_glove = [str(i) for i in range(len(dataset_evaluacion))]

    coleccion_glove.add(
        documents=documentos_glove,
        metadatas=metadatos_glove,
        ids=ids_glove
    )
    print(f"ColecciГіn 'conocimiento_bancario_glove' creada y poblada con {coleccion_glove.count()} documentos.")
else:
    print(f"ColecciГіn 'conocimiento_bancario_glove' ya contiene documentos. Saltando operaciГіn de aГ±adir. Documentos: {coleccion_glove.count()}.")

#5
from sentence_transformers import SentenceTransformer
sentences = ["This is an example sentence", "Each sentence is converted"]

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
embeddings = model.encode([item["pregunta"] for item in dataset_evaluacion])
print(embeddings)

#5
!pip install chromadb --quiet
!pip install faiss-cpu --quiet

import chromadb
from chromadb.utils import embedding_functions

# 1. Configurar el cliente y crear una colecciГіn (tu base de conocimientos)
client = chromadb.Client()
# Usamos un modelo de embedding por defecto o el que hayas elegido (ej. Sentence Transformers)
embedding_fn = embedding_functions.DefaultEmbeddingFunction()
coleccion = client.get_or_create_collection(name="conocimiento_bancario", embedding_function=embedding_fn)

# 2. Preparar los datos de tu dataset (Fase 1 y TP4)
# Extraemos las preguntas para vectorizar y las respuestas como 'metadatos'
documentos = [item["pregunta"] for item in dataset_evaluacion]
metadatos = [{"respuesta": item["respuesta"]} for item in dataset_evaluacion]
ids = [str(i) for i in range(len(dataset_evaluacion))]

# 3. Cargar el dataset en la base de datos vectorial
# Only add documents if the collection is empty to avoid duplicates on re-execution.
if coleccion.count() == 0:
    coleccion.add(
        documents=documentos,
        metadatas=metadatos,
        ids=ids
    )
else:
    print("Collection already contains documents. Skipping add operation.")

# 4. FunciГіn de bГєsqueda para tu ChatBot (Motor de bГєsqueda)
def buscar_en_db(consulta_usuario):
    # La DB vectoriza la consulta y busca la similitud del coseno automГЎticamente [8, 9]
    resultados = coleccion.query(
        query_texts=[consulta_usuario],
        n_results=1
    )
    # Retornamos la respuesta recuperada
    # The 'metadatas' field from ChromaDB query results is a list of lists.
    # We need to access the first element of the outer list, then the first element
    # of the inner list to get the metadata dictionary, and finally the 'respuesta' key.
    return resultados['metadatas'][0][0]['respuesta']

# Prueba rГЎpida
print(f"Chatbot: {buscar_en_db('ВїCГіmo recupero mi clave?')}")

# Bucle infinito para mantener la conversaciГіn activa
while True:
    # 1. Solicitar la entrada del usuario
    entrada_usuario = input("TГє: ")

    # 2. CondiciГіn de salida para cerrar el chat
    if entrada_usuario.lower() in ["salir", "chau", "exit"]:
        print("Chatbot: ВЎHasta luego! Gracias por consultar al Banco Nacional.")
        break

    # 3. Llamar a tu funciГіn de bГєsqueda con la pregunta ingresada
    # AquГ­ reemplazamos la consulta fija por 'entrada_usuario'
    respuesta = buscar_en_db(entrada_usuario)

    # 4. Mostrar la respuesta recuperada
    print(f"Chatbot: {respuesta}")

#6
class ChatBot:
    def __init__(self, llm_pipeline, db_collection):
        self.llm_pipeline = llm_pipeline
        self.db_collection = db_collection

    def responder(self, query):
        # 1. Recuperar contexto relevante de la base de datos vectorial
        # La colecciГіn de ChromaDB se encarga de embeddear la query con su funciГіn de embedding configurada
        resultados = self.db_collection.query(
            query_texts=[query],
            n_results=1
        )

        # Extraer el contexto (respuesta mГЎs similar)
        contexto = resultados['metadatas'][0][0]['respuesta']

        # 2. Generar respuesta usando el LLM y el contexto recuperado
        # Formato de prompt para el LLM
        prompt = f"Basado en la siguiente informaciГіn: '{contexto}', responde a la pregunta: '{query}'."

        # Generar la respuesta con el LLM
        # Ajustamos parГЎmetros para una generaciГіn mГЎs controlada
        response = self.llm_pipeline(
            prompt,
            max_new_tokens=50,      # Limita la longitud de la respuesta
            num_return_sequences=1, # Solo queremos una respuesta
            truncation=True         # Trunca el prompt si es demasiado largo para el modelo
        )
        return response[0]['generated_text']

# 8. InstanciaciГіn de los ChatBots para comparaciГіn (Consigna d)
# bot_st: Usa la colecciГіn 'conocimiento_bancario' (creada en la celda anterior,
#         que utiliza la DefaultEmbeddingFunction de ChromaDB, normalmente SentenceTransformer)
#         Adaptamos el nombre 'bot_spacy' a 'bot_st' para mayor claridad con el modelo de embedding usado.
bot_st = ChatBot(llm_pipeline, coleccion)

# bot_gensim: Usa la colecciГіn 'conocimiento_bancario_glove' (creada con Gensim GloVe embeddings)
bot_gensim = ChatBot(llm_pipeline, coleccion_glove)

print("ChatBots 'bot_st' y 'bot_gensim' instanciados exitosamente.")

#7
import pandas as pd

# Lista para guardar los resultados de la comparativa
resultados_comparativa = []

print("--- Iniciando EvaluaciГіn Comparativa ---")

for item in dataset_evaluacion:
    pregunta = item["pregunta"]
    ground_truth = item["respuesta"] # La respuesta real esperada

    # Obtener respuestas de ambos bots (instanciados previamente en el punto c)
    res_spacy = bot_spacy.responder(pregunta)
    res_gensim = bot_gensim.responder(pregunta)

    resultados_comparativa.append({
        "Pregunta": pregunta,
        "Respuesta Real (Dataset)": ground_truth,
        "Bot (SpaCy - md)": res_spacy,
        "Bot (Gensim - GloVe)": res_gensim
    })

# Crear DataFrame para visualizar los resultados
df_resultados = pd.DataFrame(resultados_comparativa)
display(df_resultados)
```
