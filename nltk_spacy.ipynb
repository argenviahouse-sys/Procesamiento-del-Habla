# LIMPIANDO TEXTO en PYTHON

### 1. Preprocesamiento de Texto con NLTK

NLTK (Natural Language Toolkit) es una de las librerías más clásicas para el procesamiento de texto. Primero, instalaremos las dependencias y descargaremos los recursos necesarios para el español.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer

# Descargar recursos necesarios
# 'punkt' y 'punkt_tab' son necesarios para la tokenización
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

print("Recursos de NLTK descargados correctamente.")
```

```python
# Texto de ejemplo en español
texto_espanol = "Los estudiantes están aprendiendo procesamiento de lenguaje natural en la clase de hoy."

# 1. Tokenización: Dividir el texto en palabras individuales
tokens = word_tokenize(texto_espanol.lower(), language='spanish')

# 2. Stop Words: Eliminar palabras comunes que no aportan mucho significado (ej: 'los', 'de', 'en')
lista_stop_words = set(stopwords.words('spanish'))
filtrados = [word for word in tokens if word.isalpha() and word not in lista_stop_words]

# 3. Stemming: Reducir las palabras a su raíz (usando Snowball para español)
stemmer = SnowballStemmer('spanish')
stems = [stemmer.stem(word) for word in filtrados]

print(f"Texto Original: {texto_espanol}")
print(f"Tokens: {tokens}")
print(f"Sin Stop Words: {filtrados}")
print(f"Después de Stemming: {stems}")
```

### 2. Preprocesamiento de Texto con SpaCy

SpaCy es una librería más industrial. A diferencia de NLTK, SpaCy no suele usar *stemming* (basado en reglas de recorte), sino **Lematización**, que devuelve la forma base real de la palabra (el lema) consultando un diccionario.

```python
# Instalar el modelo de español de SpaCy
!python -m spacy download es_core_news_sm
```

```python
import spacy

# Cargar el modelo en español
nlp = spacy.load('es_core_news_sm')

texto_espanol = "Los estudiantes están aprendiendo procesamiento de lenguaje natural en la clase de hoy."

# Procesar el texto
doc = nlp(texto_espanol)

# Realizar el preprocesamiento con una comprensión de lista
# Filtramos por stop words y puntuación, y obtenemos el lema
limpieza_spacy = [
    {
        "Original": token.text,
        "Lema": token.lemma_,
        "Es_Stopword": token.is_stop
    }
    for token in doc if not token.is_punct
]

print(f"Texto Original: {texto_espanol}\n")
print("Resultado detallado con SpaCy:")
for item in limpieza_spacy:
    print(f"Palabra: {item['Original']:<15} | Lema: {item['Lema']:<15} | ¿Es Stopword?: {item['Es_Stopword']}")
```

### 3. Comparativa: Stemming vs. Lematización

| Característica | Stemming (NLTK) | Lematización (SpaCy) |
| :--- | :--- | :--- |
| **Método** | Algoritmo basado en reglas (recorte). | Análisis morfológico y diccionario. |
| **Resultado** | Raíz o 'Stem' (puede no ser una palabra). | Lema (palabra real en infinitivo/singular). |
| **Precisión** | Menor (ej: 'estudi'). | Mayor (ej: 'estudiante'). |
| **Velocidad** | Muy rápido. | Un poco más lento (requiere cargar modelos). |

**¿Cuál es mejor?**
- **SpaCy** es la mejor opción para aplicaciones profesionales, chatbots y análisis de sentimientos, ya que entiende el contexto gramatical.
- **NLTK** es preferible si necesitas algo muy ligero o estás realizando tareas de recuperación de información muy básicas donde la gramática no es prioritaria.

```python
# Vamos a comparar directamente un par de palabras
palabras_prueba = ["estudiantes", "están", "aprendiendo", "perros", "cacería"]

# Stemming con NLTK
stemmer = SnowballStemmer('spanish')
resultados_stem = [stemmer.stem(p) for p in palabras_prueba]

# Lematización con SpaCy
doc_prueba = nlp(" ".join(palabras_prueba))
resultados_lema = [token.lemma_ for token in doc_prueba]

print(f"Palabras originales: {palabras_prueba}")
print(f"Stemming (NLTK):    {resultados_stem}")
print(f"Lematización (SpaCy): {resultados_lema}")
```

### 4. Tokenización: Conceptos y ejemplos

La tokenización es el proceso de segmentar el texto. Existen dos tipos principales:
- **Tokenización de oraciones**: Divide un párrafo en frases.
- **Tokenización de palabras**: Divide una frase en palabras y signos.

### 5. Preprocesamiento con Gensim

Gensim es una librería diseñada para el modelado de temas y análisis de similitud. Aunque no es su función principal, ofrece herramientas útiles para la limpieza de texto de forma muy rápida.

```python
# 1. Instalar gensim
!pip install gensim
```

```python


from gensim.utils import simple_preprocess
from nltk.corpus import stopwords
import nltk

# Aseguramos que las stopwords de NLTK estén disponibles
nltk.download('stopwords')

texto_ejemplo = "Los estudiantes están aprendiendo procesamiento de lenguaje natural en la clase de hoy."

# 2. simple_preprocess: Tokeniza, pasa a minúsculas y elimina acentos automáticamente
tokens_gensim = simple_preprocess(texto_ejemplo, deacc=True)

# 3. Filtrado de Stop Words en español
stop_words_es = set(stopwords.words('spanish'))
filtrado_gensim = [word for word in tokens_gensim if word not in stop_words_es]

print(f"Texto Original: {texto_ejemplo}")
print(f"Tokens (Gensim): {tokens_gensim}")
print(f"Filtrado (Gensim + NLTK): {filtrado_gensim}")
```

```python
texto_complejo = "¡Hola! El NLP es fascinante. ¿Están listos para programar en Python?"

print("--- Tokenización con NLTK ---")
from nltk.tokenize import sent_tokenize, word_tokenize

oraciones_nltk = sent_tokenize(texto_complejo, language='spanish')
palabras_nltk = word_tokenize(texto_complejo, language='spanish')

print(f"Oraciones ({len(oraciones_nltk)}): {oraciones_nltk}")
print(f"Palabras ({len(palabras_nltk)}): {palabras_nltk}")

print("\n--- Tokenización con SpaCy ---")
doc_spacy = nlp(texto_complejo)

# SpaCy detecta oraciones automáticamente
oraciones_spacy = [sent.text for sent in doc_spacy.sents]
# Y palabras/tokens
palabras_spacy = [token.text for token in doc_spacy]

print(f"Oraciones ({len(oraciones_spacy)}): {oraciones_spacy}")
print(f"Palabras ({len(palabras_spacy)}): {palabras_spacy}")
```

### 6. Comparativa Final: NLTK vs. SpaCy vs. Gensim

| Librería | Enfoque Principal | Fortalezas en Preprocesamiento |
| :--- | :--- | :--- |
| **NLTK** | Educación e Investigación. | Muy flexible, granular y con muchos algoritmos clásicos (Stemming). |
| **SpaCy** | Producción Industrial. | Extremadamente rápido, preciso y con Lematización avanzada 'out-of-the-box'. |
| **Gensim** | Modelado de Temas (Vectores). | Limpieza de texto muy simple y eficiente para grandes volúmenes de datos. |

**Conclusión para estudiantes:**
- Usen **SpaCy** para proyectos modernos y precisos.
- Usen **NLTK** para entender los fundamentos o cuando necesiten un control total sobre cada paso.
- Usen **Gensim** cuando el objetivo final sea crear modelos de Word Embeddings o Topic Modeling.
