# PUNTO 2

Crear el chatbot utilizando TFIDF y similitud del coseno

**Código para el Chatbot (TF-IDF + Similitud del Coseno)**

Construir un motor de búsqueda que utilice la técnica estadística TF-IDF para vectorizar las preguntas y la similitud del coseno para encontrar la mejor respuesta.

```python
# 2. Vectorización con TF-IDF
# Esta herramienta asigna un peso a cada palabra según su relevancia [4, 5]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(preguntas)

# 3. Función de respuesta
def responder_pregunta_tfidf(consulta_usuario):
    # Convertir la consulta al mismo formato numérico que el dataset [6, 7]
    consulta_vec = vectorizer.transform([consulta_usuario])

    # Calcular la similitud del coseno (valor entre 0 y 1) [8, 9]
    similitudes = cosine_similarity(consulta_vec, tfidf_matrix).flatten()

    # Encontrar la pregunta con mayor puntaje
    indice_mejor = similitudes.argmax()
    puntuacion = similitudes[indice_mejor]

    # Umbral de confianza: si es menor a 0.2, el bot no está seguro
    if puntuacion < 0.2:
        return "Lo siento, no entiendo tu consulta. ¿Podrías ser más específico?"

    return respuestas[indice_mejor]

# --- PRUEBA ---
pregunta_usuario = "¿Qué papeles necesito para un crédito?"
print(f"Pregunta: {pregunta_usuario}")
print(f"Chatbot: {responder_pregunta_tfidf(pregunta_usuario)}")
```

4- analice las fortalezas y debilidades técnicas de cada método según lo aprendido en las fuentes:
Sobre TF-IDF: es una herramienta de vectorización estadística muy eficiente para detectar palabras clave importantes, pero que tiene un defecto enorme: no mide la importancia real del contexto y pierde el orden de las palabras. Es ideal para búsquedas literales pero rígido ante el lenguaje natural.

Sobre Embeddings: este motor da un "salto a la semántica", representando las palabras como vectores densos que capturan significados y relaciones. Se destaca que su gran ventaja es entender que palabras diferentes pueden referirse al mismo concepto (como "rey" y "monarca").

Comparación de fallos: Se indica en qué casos el modelo de embeddings pudo confundir intenciones debido a la cercanía de vectores en temas similares, o por qué TF-IDF falló al no encontrar términos exactos en su diccionario.

5- Ejemplos de análisis para esta conclusión:

Ejemplo 1: El desafío de los sinónimos (Embeddings vs. TF-IDF)
Consulta: "¿Qué papeles necesito para un crédito?"
Dataset original: "¿Cuáles son los requisitos para un préstamo?"
Análisis: En este caso, el motor TF-IDF fallará o dará un puntaje muy bajo porque las palabras "papeles" y "crédito" no aparecen en la pregunta original del dataset.

En cambio, el motor de Embeddings ganará, ya que el modelo preentrenado (como es_core_news_md) reconoce que "papeles" es semánticamente cercano a "requisitos" y "crédito" a "préstamo" en un espacio vectorial.


Ejemplo 2: El problema de la importancia de las palabras (Ponderación)
Consulta: "Necesito saber el saldo de la cuenta".
Análisis: Un modelo simple de frecuencia podría darle peso a "el" y "la". Sin embargo, TF-IDF mejora esto castigando esas palabras comunes (IDF bajo) y premiando "saldo" y "cuenta" (IDF alto).

Aun así, si el usuario preguntara "¿Cuánto dinero tengo?", TF-IDF fallaría por falta de coincidencia de términos, mientras que los Embeddings triunfarían al capturar la intención de "dinero" vinculada a "saldo".

Recomendación final: SpaCy es la mejor opción para aplicaciones industriales y chatbots precisos porque entiende el contexto gramatical, mientras que TF-IDF es útil para recuperaciones más básicas donde la gramática no es prioritaria
